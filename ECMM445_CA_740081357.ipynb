DATA PREPROCESSING AND CLEANING

```python
#stripping leading and trailing spaces of column names
df.columns = df.columns.str.strip()

# Checking for NULL values
print("\nMissing Values Per Column:")
print(df.isnull().sum())
```

```python
#Checking Duplicate Values
print("Number of Duplicates in Dataset:")
print(df.duplicated().sum())
```

```python
# Removing irrelevant or redundant columns
columns_to_drop = ["EmployeeCount", "Over18", "StandardHours", "EmployeeNumber"]  # These columns add no value
df = df.drop(columns=columns_to_drop)
```

```python
# Handling categorical data (standardizing strings)
categorical_columns = df.select_dtypes(include=['object']).columns
df[categorical_columns] = df[categorical_columns].apply(lambda x: x.str.strip().str.lower())

# Checking for unique values in categorical columns to identify inconsistencies
print("\nUnique Values in Categorical Columns:")
for col in categorical_columns:
    print(f"{col}: {df[col].unique()}")
```

ENCODING CATEGORICAL VARIABLES

```python
# Binary Encoding for binary categorical variables
df['Attrition'] = df['Attrition'].map({'yes': 1, 'no': 0})
df['Gender'] = df['Gender'].map({'male': 1, 'female': 0})
df['OverTime'] = df['OverTime'].map({'yes': 1, 'no': 0})

# One-Hot Encoding for multi-category variables
df = pd.get_dummies(df, columns=['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus'], drop_first=True)

print("Columns after encoding categorical variables:")
print(df.columns)
```

FEATURE SELECTION

```python
X = df.drop(columns=["Attrition"])  # Replace "Attrition" with your target column name
y = df["Attrition"]
```

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import matplotlib.pyplot as plt

# Train Random Forest model
rf = RandomForestClassifier(n_estimators=100, random_state=5)
rf.fit(X, y)

# Get feature importance
feature_importance = rf.feature_importances_

# Create a DataFrame for features and their importance
importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importance
})

# Sort features by importance in descending order
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Plot the top features
plt.figure(figsize=(10, 6))
plt.barh(importance_df['Feature'][:15], importance_df['Importance'][:15], color='orange')
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Top 15 Features Influencing Attrition Rate')
plt.gca().invert_yaxis()  # To display the highest importance on top
plt.show()

# Print the top 10 features
print("Top 15 Features Influencing Attrition Rate:")
print(importance_df.head(15))
top_features = importance_df.head(15) 
```

IMBALANCE HANDLING --- USING SMOTE

```python
# Select top features for next steps
X_selected = X[top_features["Feature"]]

# Step 2: Handling Imbalance using SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_selected, y)

print("Original dataset shape:", X.shape)
print("Resampled dataset shape:", X_resampled.shape)
```

LOGISTIC REGRESSION

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve, precision_recall_curve, average_precision_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Logistic Regression Model
log_reg_model = LogisticRegression(max_iter=1000, random_state=42)
log_reg_model.fit(X_train, y_train)

# Predictions and evaluation
y_pred = log_reg_model.predict(X_test)
y_pred_prob = log_reg_model.predict_proba(X_test)[:, 1]

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Plot ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color="blue", label=f"Logistic Regression (AUC = {roc_auc_score(y_test, y_pred_prob):.2f})")
plt.plot([0, 1], [0, 1], color="red", linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
# Plot Confusion Matrix as Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", cbar=False, 
            xticklabels=['Predicted No Attrition', 'Predicted Attrition'], 
            yticklabels=['Actual No Attrition', 'Actual Attrition'])
plt.title("Confusion Matrix - Logistic Regression")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python

# Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_test, y_pred_prob)

# Plot Precision-Recall Curve
plt.figure(figsize=(10, 6))
plt.plot(recall, precision, color="green", label=f"Precision-Recall curve (AP = {average_precision_score(y_test, y_pred_prob):.2f})")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.show()
```

CROSS VALIDATION

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

# Perform cross-validation for both feature sets
# With 15 features
log_reg_model = LogisticRegression(max_iter=1000, random_state=42)
cv_scores = cross_val_score(log_reg_model, X_resampled, y_resampled, cv=5, scoring='accuracy')

print(f"Cross-validation scores with 15 features: {cv_scores}")
print(f"Average cross-validation score with 15 features: {cv_scores.mean()}")
print(f"Standard Deviation of CV scores: {cv_scores.std():.2f}")
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve,precision_recall_curve,auc
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.19, random_state=42)

# Random Forest Model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, max_depth=None, class_weight='balanced')
rf_model.fit(X_train, y_train)

# Predictions and evaluation
y_pred = rf_model.predict(X_test)
y_pred_prob = rf_model.predict_proba(X_test)[:, 1]

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Plot ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color="orange", label=f"Random Forest (AUC = {roc_auc_score(y_test, y_pred_prob):.2f})")
plt.plot([0, 1], [0, 1], color="red", linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Random Forest")
plt.legend()
plt.show()

# Plot Confusion Matrix as Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Oranges", cbar=False, 
            xticklabels=['Predicted No Attrition', 'Predicted Attrition'], 
            yticklabels=['Actual No Attrition', 'Actual Attrition'])
plt.title("Confusion Matrix - Random Forest")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# Compute Precision-Recall values
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_prob)

# Calculate the Area Under the Precision-Recall Curve (PR AUC)
pr_auc = auc(recall, precision)

# Plot Precision-Recall Curve
plt.figure(figsize=(10, 6))
plt.plot(recall, precision, label=f"Random Forest (PR AUC = {pr_auc:.2f})", color="darkorange")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

# Use cross-validation to assess performance for RANDOM FOREST
cv_scores = cross_val_score(rf_model, X_resampled, y_resampled, cv=5, scoring='accuracy')
print(f"Cross-validation scores: {cv_scores}")
print(f"Mean CV score: {cv_scores.mean():.2f}")
print(f"Standard Deviation of CV scores: {cv_scores.std():.2f}")
```

No Overfitting: The performance variation across folds is small (standard deviation = 0.06), and the mean accuracy is high (0.87), suggesting the model generalizes well.
Stable Performance: The small variation (0.06) indicates that your model's performance is stable across different data splits, reducing concerns about overfitting or underfitting.

```python
from sklearn.metrics import precision_recall_curve, auc

# Calculate Precision-Recall values for training data
y_train_pred_prob = rf_model.predict_proba(X_train)[:, 1]
precision_train, recall_train, _ = precision_recall_curve(y_train, y_train_pred_prob)
pr_auc_train = auc(recall_train, precision_train)

# Calculate Precision-Recall values for test data
y_test_pred_prob = rf_model.predict_proba(X_test)[:, 1]
precision_test, recall_test, _ = precision_recall_curve(y_test, y_test_pred_prob)
pr_auc_test = auc(recall_test, precision_test)

# Print PR AUC values
print(f"Training PR AUC: {pr_auc_train:.2f}")
print(f"Test PR AUC: {pr_auc_test:.2f}")
```
